```python
from transformers import pipeline

classifier = pipeline("zero-shot-classification",
                      model = "facebook/bart-large-mnli")
```

```python
sequence_to_classify = "one day I will go to Africa and go to a party"


_labels = ['travel',
           'cooking',
           'dancing']

classifier(sequence_to_classify, _labels)
```

```python
tr_labels = ['travel',
                    'cooking',
                    'dancing',
                    'exploration']


classifier(sequence_to_classify, 
           
           tr_labels, 

           multi_label = True)
```

```python
c_sequence_to_classify = "A lion is a big prey"


party_labels = ['science',
                'politics',
                'history']

classifier(c_sequence_to_classify,
           party_labels)
```
